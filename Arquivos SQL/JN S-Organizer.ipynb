# Relatorio2

### Objetivo: Selecionar prestadores e suas respectivas agendas realizadas no dia 01/10/2012

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""
                            select agenda.data, agenda.prestador_id_usuario as prestador, solicitacao.protocolo
                            from agenda 
                            inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
                            inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)
                            WHERE agenda.data = '2012-10-01'
                            """,conn)
```

```python
result
```

## Relatório 3

### Objetivo: Selecionar um número específico de prestadores que atendem solicitações anteriores a 2013, bem como suas empresas contratantes.

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""
                            select agenda.empresa_id_usuario as empresa,
                            agenda.prestador_id_usuario as prestador,
                            solicitacao.protocolo as protocolo
                            from agenda 
                            inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
                            inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)
                            where agenda.data < '2013-01-01' limit 10;

                            """,conn)
```

```python
result
```

## Relatório 4

### Objetivo: Visualizar os prestadores que atenderam a uma faixa específica de ID’s de solicitações

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""
                            select agenda.prestador_id_usuario as prestadorDisponivel,
                            solicitacao.protocolo as SolicitacoesPendentes
                            from agenda
                            inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
                            inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)
                            inner join prestador_agenda on(agenda.id = prestador_agenda.agenda_id)
                            inner join prestador on(prestador_agenda.prestador_id_usuario = prestador.id_usuario)
                            where solicitacao.estado = 1 and prestador.disponibilidade = true
                            """,conn)
```

```python
result
```

## Relatório 5

### Objetivo:  gerar informações relativas ao serviço em agenda em uma data específica

```python
result = pd.read_sql_query("""
                    Select servico.valor, servico.categoria_id, servico.descricao, servico.horariomarcado from servico
                    inner join solicitacao_servico on(solicitacao_servico.servicos_id = servico.id)
                    inner join solicitacao on(solicitacao_servico.solicitacao_protocolo = solicitacao.protocolo)
                    inner join agenda_solicitacao on(solicitacao.protocolo = agenda_solicitacao.solicitacaododia_protocolo)
                    inner join agenda on(agenda.id = agenda_solicitacao.agenda_id)
                    Where agenda.data = '2015-03-06' 
                            """,conn)
```
