### Objetivo: Buscar a agenda de 03/05/2018 de prestadores e quais solicitações estão pendentes

Código para obtenção do resultado:

```python
test = pd.read_sql_query("""

select prestador.nome, servico.descricao, servico.horariomarcado, servico.valor, categoriaservico.tipo as servico_a_prestar from prestador

inner join prestador_agenda on(prestador.id_usuario = prestador_agenda.prestador_id_usuario)
inner join agenda on(agenda.id = prestador_agenda.agenda_id)

inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)

inner join solicitacao_servico on(solicitacao.protocolo = solicitacao_servico.solicitacao_protocolo)
inner join servico on(servico.id = solicitacao_servico.servicos_id)

inner join categoriaservico on (servico.categoria_id = categoriaservico.id)

where agenda.data = '2018-05-03' and solicitacao.estado = 0

""",conn)
```

# Relatorio2

### Objetivo: Selecionar quais prestadores concluiram um atendimento em uma certa região e o dia 

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""

select prestador.nome as nome_prestador, agenda.data as dia, endereco.CEP from prestador

inner join prestador_agenda on(prestador.id_usuario = prestador_agenda.prestador_id_usuario)
inner join agenda on(agenda.id = prestador_agenda.agenda_id)

inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)

inner join endereco on(endereco.id = solicitacao.endereco_id)

inner join atendimento on(atendimento.prestador_id_usuario = prestador.id_usuario)

where atendimento.concluido = true and endereco.CEP like '2906%'

""",conn)
```

```python
result
```

## Relatório 3

### Objetivo: Selecionar os funcionarios que atenderam e não concluiram o chamado em uma região

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""

select DISTINCT prestador.nome as nome_prestador, agenda.data as dia_do_ocorrido, categoriaservico.tipo as servico_a_prestar, servico.valor from prestador

inner join prestador_agenda on(prestador.id_usuario = prestador_agenda.prestador_id_usuario)
inner join agenda on(agenda.id = prestador_agenda.agenda_id)

inner join agenda_solicitacao on(agenda.id = agenda_solicitacao.agenda_id)
inner join solicitacao on(agenda_solicitacao.solicitacaododia_protocolo = solicitacao.protocolo)

inner join solicitacao_servico on(solicitacao.protocolo = solicitacao_servico.solicitacao_protocolo)
inner join servico on(servico.id = solicitacao_servico.servicos_id)

inner join categoriaservico on (servico.categoria_id = categoriaservico.id)

inner join atendimento on(atendimento.prestador_id_usuario = prestador.id_usuario)

inner join endereco on(endereco.id = solicitacao.endereco_id)

where solicitacao.estado = 3 and atendimento.concluido = false and endereco.CEP like '2906%'

                            """,conn)
```

```python
result
```

## Relatório 4

### Objetivo: Visualizar os lucros obtidos das empresas em 2017

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""

select empresa.nome as nome_empresa, SUM(servico.valor) AS valor_total_ganho  from empresa

inner join empresa_solicitacao on(empresa.id_usuario = empresa_solicitacao.empresa_id_usuario)
inner join solicitacao on(empresa_solicitacao.solicitacaoes_protocolo = solicitacao.protocolo)

inner join solicitacao_servico on(solicitacao.protocolo = solicitacao_servico.solicitacao_protocolo)
inner join servico on(servico.id = solicitacao_servico.servicos_id)

inner join servico_atendimento on(servico.id = servico_atendimento.servico_id)
inner join atendimento on(servico_atendimento.atendimento_id = atendimento.id)

where atendimento.concluido = true and CAST (solicitacao.datasolicitacao as varchar) like '2017%'
group by empresa.nome
                            """,conn)
```

```python
result
```

## Relatório 5

### Objetivo:  Selecionar media de multas para clientes

```python
result = pd.read_sql_query("""

select cliente.nome as nome_cliente, SUM(servico.multa) AS multa_total from cliente 

inner join cliente_solicitacao on(cliente.id_usuario = cliente_solicitacao.cliente_id_usuario)
inner join solicitacao on(cliente_solicitacao.solicitacao_protocolo = solicitacao.protocolo)

inner join solicitacao_servico on(solicitacao.protocolo = solicitacao_servico.solicitacao_protocolo)
inner join servico on(servico.id = solicitacao_servico.servicos_id)

group by cliente.nome
having AVG(servico.multa)>0

""",conn)
```

```python
result
```
